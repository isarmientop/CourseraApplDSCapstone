```python
# importar librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Datos
models = ['LR', 'SVM', 'Decision Tree', 'K Nearest']
accuracies = [0.8462429, 0.848214, 0.901785, 0.848214]
```

```python
# Crear el DataFrame
df = pd.DataFrame({
    'Model': models,
    'Accuracy': accuracies
})
```

```python
# Mostrar el DataFrame
print(df)
```

```python
# Crear el gr√°fico de barras
plt.figure(figsize=(10, 6))
plt.title('Model Accuracy Comparison')
sns.barplot(x='Model', y='Accuracy', data=df, palette='viridis')
```

```python
# A√±adir t√≠tulos y etiquetas
plt.title('Model Accuracy Comparison')
plt.xlabel('Model')
plt.ylabel('Accuracy')
```

```python
# Mostrar el gr√°fico
plt.show()
```
